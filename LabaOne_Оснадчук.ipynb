```python
def decimal_to_binary(number):
    
    if isinstance(number, int):
        if number == 0:
            return '0'
        binary = ''
        is_negative = False
        if number < 0:
            is_negative = True
            number = abs(number)
        while number > 0:
            remainder = number % 2
            binary = str(remainder) + binary
            number //= 2
        if is_negative:
            return '1|' + binary
        return '0|' + binary
    
    elif isinstance(number, float):
        integer_part = abs(int(number))
        decimal_part = abs(number) - integer_part
        integer_binary = decimal_to_binary(integer_part)
        decimal_binary = ''
        precision = 6
        for _ in range(precision):
            decimal_part *= 2
            if int(decimal_part) == 1:
                decimal_binary += '1'
                decimal_part -= int(decimal_part)
            else:
                decimal_binary += '0'
        return integer_binary + '.' + decimal_binary.rstrip('0')

print(decimal_to_binary(32))    # Виведе: 0|100000
print(decimal_to_binary(-10.25))  # Виведе: 1|1010.01
```
